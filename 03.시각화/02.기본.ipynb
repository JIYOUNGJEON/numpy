### 1. 라인 플롯

```python
plt.plot([1, 4, 9, 16])     #x=[0,1,2,3]을 가정
plt.show()
```

```python
plt.plot([1,2,3,4],[1,3,9,16])
plt.title('y=x**2 그래프')
plt.grid()
plt.show()
```

### -스타일 지정

```python
plt.plot([1,2,3,4],[1,4,9,16],'rs--')
plt.title("'rs--'스타일")
plt.grid()
plt.show()
#rs는 red square --모양 선
#색깔(color), 마커(marker), 선 종류(line style)
```

```python
plt.plot([1,2,3,4],[1,4,9,16],c='b',lw = 5,ls = "--", marker = "o",ms = 5, mec = "g", mew = 5, mfc = "r")
plt.title("스타일 적용 예")
plt.xlim(0,5)
plt.ylim(-2,20)
plt.grid
plt.show()
```

### - 틱 설정

```python
import matplotlib as mpl
mpl.rcParams['axes.unicode_minus']= False
```

```python
x = np.linspace(-np.pi, np.pi, 300)
y = np.cos(x)
plt.plot(x,y)
plt.grid()
plt.show()
```

```python
x = np.linspace(-np.pi, np.pi, 300)
y = np.cos(x)
plt.plot(x,y)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
plt.yticks([-1,0,1])
plt.title('x축과 y축의 tick label설정')
plt.grid()
plt.show()
```

```python
x = np.linspace(-np.pi, np.pi, 300)
y = np.cos(x)
plt.plot(x,y)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, 1], ["Low", "Zero", "High"])
plt.title('x축과 y축의 tick label설정')
plt.grid()      #격자선
plt.show()
```

```python
t = np.arange(0., 5., 0.2)
plt.title("라인 플롯에서 여러개의 선 그리기")
plt.plot(t, t, 'r--', t, 0.5 * t**2, 'bs:', t, 0.2 * t**3, 'g^-')
plt.grid()
plt.show()
```

```python
t = np.arange(0., 5., 0.2)
plt.plot(t,t,"r--")
plt.plot(t, 0.5 * t**2)
plt.plot(t, 0.2 * t**3, 'g^-')
plt.title("라인 플롯에서 여러개의 선 그리기")
plt.grid()
plt.show()
```

### -범례

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.title("legend를 표시한 플롯")
plt.plot(X, C, label="cosine")
plt.plot(X, S, label="sine")
plt.legend(loc=2)
plt.grid()
plt.show()
```

### -x축 y축 라벨

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.title("legend를 표시한 플롯")
plt.plot(X, C, label="cosine")
plt.plot(X, S, label="sine")
plt.legend(loc=2)
plt.xlabel("x축 라벨"); plt.ylabel('y축 라벨')
plt.grid()
plt.show()
```
