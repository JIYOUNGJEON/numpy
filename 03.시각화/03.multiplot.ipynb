```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.rcParams["font.family"] = 'Malgun Gothic'
mpl.rcParams['axes.unicode_minus']= False
```

### -figure 객체

```python
np.random.seed(2021)
f1 = plt.figure(figsize = (10,2))
plt.title("figure size : (10,2)")
plt.plot(np.random.randn(100))
plt.show
plt.grid()
```

### -Axes 객체와 subplot명령

```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)
```

```python
ax1 = plt.subplot(2,1,1)    #행이 2개 열이 1개 figure 1개
plt.title('2subplots')
ax1.plot(x1, y1, 'yo-')
ax1.set_ylabel("Damped oscillation")

ax2 = plt.subplot(2,1,2)
ax2.plot(x2,y2,'r.-')
ax2.set_ylabel("Undamped")

plt.xlabel('time (sec)')
plt.show
```

```python
ax1 = plt.subplot(2,1,1)    #행이 2개 열이 1개 figure 1개
plt.plot(x1, y1, 'yo-')
plt.title('2subplots')
plt.ylabel("Damped oscillation")

ax2 = plt.subplot(2,1,2)
plt.plot(x2,y2,'r.-')
plt.ylabel("Undamped")
plt.xlabel("time (sec)")

plt.tight_layout()
plt.show
```

```python
np.random.seed(0)

plt.subplot(221)
plt.plot(np.random.rand(5))
plt.title("axes 1")

plt.subplot(222)
plt.plot(np.random.rand(5))
plt.title("axes 2")

plt.subplot(223)
plt.plot(np.random.rand(5))
plt.title("axes 3")

plt.subplot(224)
plt.plot(np.random.rand(5))
plt.title("axes 4")

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2,2)

np.random.seed(0)
axes[0, 0].plot(np.random.rand(5))
axes[0, 0].set_title("axes 1")
axes[0, 1].plot(np.random.rand(5))
axes[0, 1].set_title("axes 2")
axes[1, 0].plot(np.random.rand(5))
axes[1, 0].set_title("axes 3")
axes[1, 1].plot(np.random.rand(5))
axes[1, 1].set_title("axes 4")

plt.tight_layout()
plt.show()
```

```python
np.random.seed(0)

fig, axes = plt.subplots(2,2)
for i in range(4):
    ax = axes[i//2,i%2] # (0,0), (0,1), (1,0), (1,1)    #행인덱스, 열인덱스
    ax.plot(np.random.rand(5))
    ax.set_title(f"axes{i+1}")
    ax.grid()

plt.tight_layout()
plt.show
```
